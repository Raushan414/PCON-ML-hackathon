```python
train_df1 = train_df.copy()
test_df1 = test_df.copy()
```

```python
import category_encoders as ce
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan','contact', 'month', 'day_of_week','poutcome']
```

```python
ohe = ce.OneHotEncoder(cols=cat_cols)
combine_df1 = ohe.fit_transform(combine_df)
train_len = train_df.shape[0]
train_df1 = combine_df1.iloc[:train_len]
test_df1 = combine_df1.iloc[train_len:]
```

```python
train_df1.head()
```

```python
train_df1['y'] = train_df['y'].apply(lambda x: 0 if x=='no' else 1)
```

```python
train_df1
```

```python
target = train_df1['y']
train_df1.drop(['ID','y'],inplace=True,axis=1)
test_df1.drop(['ID','y'],inplace=True,axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(train_df1,target)
y_pred = logreg.predict(test_df1)
df = pd.DataFrame({'Id':id,'Predicted':y_pred})
df['Predicted'] = df['Predicted'].apply(lambda x: 'no' if x==0 else 'yes')
df.to_csv('sub2.csv',index=False)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(train_df1,target)
y_pred = gbc.predict(test_df1)
df = pd.DataFrame({'Id':id,'Predicted':y_pred})
df['Predicted'] = df['Predicted'].apply(lambda x: 'no' if x==0 else 'yes')
df.to_csv('sub3.csv',index=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(train_df1,target)
y_pred = rfc.predict(test_df1)
df = pd.DataFrame({'Id':id,'Predicted':y_pred})
df['Predicted'] = df['Predicted'].apply(lambda x: 'no' if x==0 else 'yes')
df.to_csv('sub3.csv',index=False)
```

```python
#define a function that takes train and test data 
#using stratifiedk-fold cross validation since the class is imbalance and print the cross val f1_score
#also printing the confusion matrix and then return the cross val scroe

def evaluation(train_data,test_data,params):
    y = train_data['y']
    test_id = test_data['ID']
    X = train_data.drop(['ID','y'],axis=1)
    X_test = test_data.drop(['ID'],axis=1)
    
    kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=params['seed'])
    cv_score =[]
    tr_score = []
    i=1
    for train_index,test_index in kf.split(X,y):
        print('{} of KFold {}'.format(i,kf.n_splits))
        x_tr,x_vl = X.loc[train_index],X.loc[test_index]
        y_tr,y_vl = y.loc[train_index],y.loc[test_index]

        #model
        gbc = GradientBoostingClassifier(n_estimators=params['n_estimators'],learning_rate=params['learning_rate'],max_depth=params['max_depth'],min_samples_split=params['min_sample_split'])
        gbc.fit(x_tr,y_tr)

        score = f1_score(y_vl,gbc.predict(x_vl),average='weighted')
        tr_sc = f1_score(y_tr,gbc.predict(x_tr),average='weighted')
        print(confusion_matrix(y_vl,gbc.predict(x_vl)))
        print('val_f1_score:',score)
        print('train_f1_score: ',tr_sc)
        cv_score.append(score)
        tr_score.append(tr_sc)
        i+=1
    y_pred = gbc.predict(X_test)
    df = pd.DataFrame({'Id':test_id,'Predicted':y_pred})
    df['Predicted'] = df['Predicted'].apply(lambda x: 'no' if x==0 else 'yes')
    df.to_csv('temp_sub.csv',index=False)
    return cv_score
```

```python
#define a function that takes train and test data 
#using stratifiedk-fold cross validation since the class is imbalance and print the cross val f1_score
#also printing the confusion matrix and then return the cross val scroe

def evaluation(train_data,test_data,params):
    y = train_data['y']
    test_id = test_data['ID']
    X = train_data.drop(['ID','y'],axis=1)
    X_test = test_data.drop(['ID'],axis=1)
    
    cv_score =[]
    tr_score = []
    for i in range(1,6):
        print('{} of KFold'.format(i))
        x_tr,x_vl,y_tr,y_vl = train_test_split(X,y,test_size=0.1,shuffle=True,stratify=y)

        #model
        gbc = GradientBoostingClassifier(n_estimators=params['n_estimators'],learning_rate=params['learning_rate'],max_depth=params['max_depth'],min_samples_split=params['min_sample_split'])
        gbc.fit(x_tr,y_tr)

        score = f1_score(y_vl,gbc.predict(x_vl),average='weighted')
        tr_sc = f1_score(y_tr,gbc.predict(x_tr),average='weighted')
        print(confusion_matrix(y_vl,gbc.predict(x_vl)))
        print('val_f1_score:',score)
        print('train_f1_score: ',tr_sc)
        cv_score.append(score)
        tr_score.append(tr_sc)
    y_pred = gbc.predict(X_test)
    df = pd.DataFrame({'Id':test_id,'Predicted':y_pred})
    df['Predicted'] = df['Predicted'].apply(lambda x: 'no' if x==0 else 'yes')
    df.to_csv('temp_sub2.csv',index=False)
    return cv_score
```

```python
#define a function that takes train and test data 
#using stratifiedk-fold cross validation since the class is imbalance and print the cross val f1_score
#also printing the confusion matrix and then return the cross val scroe

def evaluation(train_data,test_data,params):
    y = train_data['y']
    test_id = test_data['ID']
    X = train_data.drop(['ID','y'],axis=1)
    X_test = test_data.drop(['ID'],axis=1)
    
    cv_score =[]
    tr_score = []
    for i in range(1,6):
        print('{} of KFold'.format(i))
        x_tr,x_vl,y_tr,y_vl = train_test_split(X,y,test_size=0.1,shuffle=True,stratify=y,random)

        #model
        gbc = HistGradientBoostingClassifier(learning_rate=params['learning_rate'],max_depth=params['max_depth'])
        gbc.fit(x_tr,y_tr)

        score = f1_score(y_vl,gbc.predict(x_vl),average='weighted')
        tr_sc = f1_score(y_tr,gbc.predict(x_tr),average='weighted')
        print(confusion_matrix(y_vl,gbc.predict(x_vl)))
        print('val_f1_score:',score)
        print('train_f1_score: ',tr_sc)
        cv_score.append(score)
        tr_score.append(tr_sc)
    y_pred = gbc.predict(X_test)
    df = pd.DataFrame({'Id':test_id,'Predicted':y_pred})
    df['Predicted'] = df['Predicted'].apply(lambda x: 'no' if x==0 else 'yes')
    df.to_csv('temp_sub2.csv',index=False)
    return cv_score
```

```python
params = {
    'max_depth':7,
    'seed':5,
    'min_sample_split':3,
    'n_estimators':1000,
    'learning_rate':0.08
}
score = evaluation(train_df1,test_df1,params)
np.mean(score)

#score = 0.92156
```
